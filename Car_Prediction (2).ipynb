```python
x=df.drop(columns='Price')
y=df['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer

from sklearn.pipeline import make_pipeline
```

```python
st=StandardScaler()
st.fit(x[['year','kms_driven']])
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                     (StandardScaler(),['year','kms_driven']),
                                     remainder='passthrough')
```

```python

x_train_tranform=column_trans.fit_transform(x_train)
x_test_transform=column_trans.transform(x_test)
```

# Modeling

```python

# Linear Regression Models
from sklearn.linear_model import LinearRegression

# Tree-Based Regression Models
from sklearn.tree import DecisionTreeRegressor

# Distance / Kernel-Based Models
from sklearn.neighbors import KNeighborsRegressor


#Boosting
from lightgbm import LGBMRegressor
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
knn=KNeighborsRegressor()
lgbm=LGBMRegressor()
```

```python

regressors={
    "LR":lr,
    "DT":dt,
    "KNN":knn,
    "LGBM":lgbm
}
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def train_regressor(regressor, x_train, y_train, x_test, y_test, name):
  # Train the regessor
  regressor.fit(x_train, y_train)

  # make prediction
  y_pred=regressor.predict(x_test)

  # Evaluate the regressor
  mse=mean_squared_error(y_test,y_pred)
  mae=mean_absolute_error(y_test,y_pred)
  r2=r2_score(y_test,y_pred)

  return mse, mae, r2
```

```python
# Lists to store mse, mae, r2
mse_scores = []
mae_scores = []
r2_scores = []

# Loop through each regressor
for name, reg in regressors.items():

    current_mse, current_mae, current_r2 = train_regressor(
        reg, x_train_tranform, y_train, x_test_transform, y_test, name
    )

    # Print results
    print(f'Name: {name}')
    print(f'MSE: {current_mse:.4f}')
    print(f'MAE: {current_mae:.4f}')
    print(f'R2: {current_r2:.4f}')
    print('----------------------------')

    # Store results
    mse_scores.append(current_mse)
    mae_scores.append(current_mae)
    r2_scores.append(current_r2)
```

```python
pipe1=make_pipeline(column_trans,lr)
pipe2=make_pipeline(column_trans,dt)
pipe3=make_pipeline(column_trans,knn)
pipe4=make_pipeline(column_trans,lgbm)
```

```python
pipe1.fit(x_train,y_train)
pipe2.fit(x_train,y_train)
pipe3.fit(x_train,y_train)
pipe4.fit(x_train,y_train)
```

```python
y_pred1=pipe1.predict(x_test)
y_pred2=pipe2.predict(x_test)
y_pred3=pipe3.predict(x_test)
y_pred4=pipe4.predict(x_test)
```

```python
print(r2_score(y_test,y_pred1))
print(r2_score(y_test,y_pred2))
print(r2_score(y_test,y_pred3))
print(r2_score(y_test,y_pred4))
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    dt = DecisionTreeRegressor()
    pipe=make_pipeline(column_trans,dt)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```
